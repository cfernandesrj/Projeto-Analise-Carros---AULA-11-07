```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("vendas_loja_seminovos.csv", sep= ';')
df
```

```python
df_dados = df[['Idade', 'Estado Civil', 'Cidade', 'Estado', 'Marca', 'Modelo', 'Ano_fabricacao', 'Cor', 'Valor_venda', 'Data_venda', 'Forma_pagamento' ]]
df_dados['Data_venda'] = pd.to_datetime(df['Data_venda'])
df_dados['ano_venda'] = df_dados['Data_venda'].dt.year
df_dados
```

```python
df_ano = df_dados.groupby(['Modelo']).size().reset_index(name= 'Vendidos')
df_ano
```

```python
df['Data_venda'] = pd.to_datetime(df['Data_venda'])
df['Ano_venda'] = df['Data_venda'].dt.year
#adicionando a aba quantidade p verificar quantos carros da marca foi vendido
vendas_marca_ano = df.groupby(['Ano_venda', 'Marca']).size().reset_index(name='Quantidade')
#organizando por qantidade excluindo os anos repetidos e fica so ano e quantidade
marca_top_ano = vendas_marca_ano.sort_values('Quantidade', ascending=False).drop_duplicates(['Ano_venda'])
marca_top_ano
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=marca_top_ano, x='Ano_venda', y='Quantidade', hue='Marca')
plt.title('Marca mais vendida por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Vendas')
plt.legend(title='Marca')
plt.tight_layout()
plt.show()
```

```python
resultados = []
for ano in df_dados['ano_venda'].unique():
    df_ano = df_dados[df_dados['ano_venda'] == ano]
    marca_top = df_ano['Marca'].value_counts().idxmax()
    df_marca_top = df_ano[df_ano['Marca'] == marca_top]
    modelo_top = df_marca_top['Modelo'].value_counts().idxmax()
    quantidade = df_marca_top['Modelo'].value_counts().max()
    resultados.append({
        'ano_venda': ano,
        'Marca': marca_top,
        'Modelo': modelo_top,
        'quantidade': quantidade
    })

df_resultados = pd.DataFrame(resultados)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_resultados, x='ano_venda', y='quantidade', hue='Modelo')
plt.title('Modelo mais vendido da marca com mais vendas por ano')
plt.xlabel('Ano de Venda')
plt.ylabel('Quantidade de Vendas')
plt.legend(title='Modelo')
plt.tight_layout()
plt.show()
```
