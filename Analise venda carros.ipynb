```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("vendas_loja_seminovos.csv", sep= ';')
df
```

```python
df_dados = df[['Idade', 'Estado Civil', 'Cidade', 'Estado', 'Marca', 'Modelo', 'Ano_fabricacao', 'Cor', 'Valor_venda', 'Data_venda', 'Forma_pagamento' ]]
df_dados['Data_venda'] = pd.to_datetime(df['Data_venda'])
df_dados['ano_venda'] = df_dados['Data_venda'].dt.year
df_dados
```

```python
df_ano = df_dados.groupby(['Modelo']).size().reset_index(name= 'Vendidos')
df_ano
```

```python
df['Data_venda'] = pd.to_datetime(df['Data_venda'])
df['Ano_venda'] = df['Data_venda'].dt.year
vendas_marca_ano = df.groupby(['Ano_venda', 'Marca']).size().reset_index(name='Quantidade')
vendas_marca_ano
```

```python
marca_top_ano = vendas_marca_ano.sort_values('Quantidade', ascending=False).drop_duplicates(['Ano_venda'])
marca_top_ano
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=marca_top_ano, x='Ano_venda', y='Quantidade', hue='Marca')
plt.title('Marca mais vendida por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Vendas')
plt.legend(title='Marca')
plt.tight_layout()
plt.show()
```

```python
# Passo 2: Marca mais vendida por ano
vendas_marca = df.groupby(['Ano_venda', 'Marca']).size().reset_index(name='Qtd_vendas')
marca_top = vendas_marca.sort_values('Qtd_vendas', ascending=False).drop_duplicates('Ano_venda')

# Passo 3: Filtrar o DataFrame apenas com as marcas campeãs por ano
df_top_marcas = pd.merge(df, marca_top[['Ano_venda', 'Marca']], on=['Ano_venda', 'Marca'])

# Passo 4: Encontrar o modelo mais vendido de cada uma dessas marcas por ano
vendas_modelo = df_top_marcas.groupby(['Ano_venda', 'Modelo']).size().reset_index(name='Qtd_vendas')
modelo_top = vendas_modelo.sort_values('Qtd_vendas', ascending=False).drop_duplicates('Ano_venda')

# Passo 5: Juntar marca e modelo (pra usar no gráfico)
df_final = pd.merge(marca_top, modelo_top, on='Ano_venda', suffixes=('_marca', '_modelo'))

# Agora temos: Ano_venda | Marca | Qtd_vendas_marca | Modelo | Qtd_vendas_modelo

# Passo 6: Gráfico com Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_final, x='Ano_venda', y='Qtd_vendas_marca', hue='Modelo_modelo', palette='Set2')

# Adiciona o nome da marca no topo da barra
for index, row in df_final.iterrows():
    plt.text(x=row['Ano_venda'], 
             y=row['Qtd_vendas_marca'] + 5, 
             s=row['Marca'], 
             ha='center', fontsize=9, fontweight='bold')

plt.title('Marca + Modelo mais vendidos por ano')
plt.xlabel('Ano da Venda')
plt.ylabel('Quantidade de Vendas')
plt.legend(title='Modelo mais vendido da marca campeã')
plt.tight_layout()
plt.show()
```
